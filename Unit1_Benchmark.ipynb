Experiment 1: Text Generation
Task: Try to generate text using the prompt: "The future of Artificial Intelligence is"

BERT (bert-base-uncased): An Encoder-only model

```python
generator = pipeline('text-generation', model='bert-base-uncased')
output = generator("The future of Artificial Intelligence is", max_length=50, num_return_sequences=1)
print(output[0]['generated_text'])
```

RoBERTa (roberta-base): An optimized Encoder-only model.

```python
generator = pipeline('text-generation', model='roberta-base')
output = generator("The future of Artificial Intelligence is", max_length=50, num_return_sequences=1)
print(output[0]['generated_text'])
```

BART (facebook/bart-base): An Encoder-Decoder model

```python
generator = pipeline('text-generation', model='facebook/bart-base')
output = generator("The future of Artificial Intelligence is", max_length=50, num_return_sequences=1)
print(output[0]['generated_text'])
```

Experiment 2: Masked Language Modeling (Missing Word)

```python
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
fill_mask("The goal of Generative AI is to [MASK] new content.")
```

```python
fill_mask = pipeline("fill-mask", model="roberta-base")
fill_mask("The goal of Generative AI is to <mask> new content.")
```

```python
fill_mask = pipeline("fill-mask", model="facebook/bart-base")
fill_mask("The goal of Generative AI is to <mask> new content.")
```

Experiment 3: Question Answering

```python
qa = pipeline("question-answering", model="bert-base-uncased")

qa({
    "question": "What are the risks?",
    "context": "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
})
```

```python
qa = pipeline("question-answering", model="roberta-base")

qa({
    "question": "What are the risks?",
    "context": "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
})
```

```python
qa = pipeline("question-answering", model="facebook/bart-base")

qa({
    "question": "What are the risks?",
    "context": "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
})
```
